```python

from variables import MONGO_URL
from pymongo import MongoClient
from fixture_formatter import FixtureFormatter
import datetime
import sys
import time
import threading
from sportmonks import SportMonksAPI
import certifi
ca = certifi.where()
client = MongoClient(MONGO_URL, tlsCAFile=ca)

nan = None
api = SportMonksAPI()
database = client.footyamigo
Fixture = database.fixtures


class LiveBot:
    def __init__(self):
        self.MAX_WAIT_SECONDS = 20
        self.CHUNK_SIZE = 50

    @staticmethod
    def upload_fixture(fixture):
        try:
            formatter = FixtureFormatter(is_live=True)
            formatted_fixture = eval(str(formatter.format(fixture)))
            fixture_id = fixture.get("id")
            Fixture.update_one(
                {'fixture_id': fixture_id}, formatted_fixture, upsert=True)
            response = {
                "statusCode": 200,
                "body": f"Fixture ID: {fixture_id} upload success"
            }
            sys.stdout.write('\r'+str(fixture["id"]) + " success")
            return response
        except Exception as e:
            print("\nError occured:", fixture["id"], e, "\n")

    @staticmethod
    def perform_async_tasks(items, target):
        tasks = []
        for item in items:
            t = threading.Thread(target=target,
                                 args=(item,))
            tasks.append(t)
            t.start()
        for t in tasks:
            t.join()

    def upload_fixtures_async(self, fixture_ids):
        fixtures = api.get_fixtures_by_ids(fixture_ids)
        self.perform_async_tasks(fixtures, self.upload_fixture)

    def get_fixtures_async(self):
        fixture_ids = api.get_live_fixture_ids()
        chunks = [fixture_ids[i:i + self.CHUNK_SIZE]
                  for i in range(0, len(fixture_ids), self.CHUNK_SIZE)]
        self.perform_async_tasks(chunks, self.upload_fixtures_async)

    def start_bot(self):
        while True:
            start = time.time()
            try:
                self.get_fixtures_async()
            except Exception as e:
                print(e)
            
            end = time.time()
            ELAPSED = round(end - start)
            WAIT_SECONDS = max(0, self.MAX_WAIT_SECONDS-ELAPSED)
            print("\nELAPSED",ELAPSED, "SECONDS", ":","WAITING", self.MAX_WAIT_SECONDS, "SECONDS")
            time.sleep(WAIT_SECONDS)
bot = LiveBot()
bot.start_bot()
```
