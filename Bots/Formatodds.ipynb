```python
import json
import pandas as pd
```

```python
with open("fixture.json") as f:
  fixture = json.loads(f.read())
```

```python
fixture['data']['odds']['data']
```

```python
odds_data = []
for market in fixture['data']['odds']['data']:
    market_name = market['name']
    market_id = market['id']
    for bookmaker in market['bookmaker']['data']:
        bookmaker_name = bookmaker['name']
        bookmaker_id = bookmaker['id']
        if bookmaker_id != 2:
            continue
        bookmaker_odds = [{"market_name": market_name, "market_id": market_id,
                         "bookmaker_name": bookmaker_name, "bookmaker_id": bookmaker_id, **odd} for odd in bookmaker['odds']['data']]
        odds_data.extend(bookmaker_odds)
```

```python
pd.DataFrame(odds_data).to_csv("odds_data.csv")
```
